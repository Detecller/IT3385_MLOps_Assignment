- The correlation matrix does not reveal strong inter-feature correlations, apart from those involving the diagnosis variable.
#### Observed relationships are consistent with established clinical understanding of Alzheimer’s disease
- Positive correlation between Alzheimer's diagnosis with memory complaints (0.31), behavioural problems (0.22)
- Negative correlation between Alzheimer's diagnosis with ADL (Activities of Daily Living) (-0.33), FunctionalAssessment (-0.36) and MMSE (Mini-Mental State Examination score) (-0.24)

# ML Pipeline with PyCaret

```python
import mlflow
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
# Initialise experiement
from pycaret.classification import *
exp = ClassificationExperiment()
```

```python
# Define columns based on their type
nominal_cols = ['Gender', 'Ethnicity', 'Smoking', 'FamilyHistoryAlzheimers', 
            'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury', 
            'Hypertension', 'MemoryComplaints', 'BehavioralProblems', 
            'Confusion', 'Disorientation', 'PersonalityChanges', 
            'DifficultyCompletingTasks', 'Forgetfulness']

continuous_cols = ['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 
              'SleepQuality', 'SystolicBP', 'DiastolicBP', 'CholesterolTotal', 
              'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 
              'MMSE', 'FunctionalAssessment', 'ADL']
```

```python
exp.setup(
    df,
    target='Diagnosis',
    categorical_features = nominal_cols,
    numeric_features = continuous_cols,
    ordinal_features = {'EducationLevel' : [0, 1, 2, 3]},
    ignore_features=['DoctorInCharge', 'PatientID'],
    normalize=True,
    normalize_method = 'zscore',
    session_id=45,
    log_experiment=True,
    experiment_name="alzheimer_exp"
)
```

```python
best_model = exp.compare_models(sort='Recall')
```

```python
tuned_model = exp.tune_model(
    best_model,
    optimize='Recall',
    search_library = 'optuna',
    n_iter = 5
)
```

```python
results = exp.predict_model(tuned_model)
```

- Similar results suggest the model can generalise well on unseen data
